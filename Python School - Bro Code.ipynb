# 24 . *args

```python
# *args = parameter that will pack all arguments into a tuple.
#         useful so that a function can accept a varying amount of arguments

def add(num1, num2):
    sum = num1 + num2
    return sum
print(add(1, 2)) # more than 2 arguments returns Error


def add(*stuff):
    sum = 0
    for i in stuff:
        sum = sum + i
    return sum
print(add(1, 2, 3, 4, 5, 6)) # returns true value


# if you need to change value of argument
def add(*stuff):
    sum = 0
    stuff = list(stuff) # tuples and not changeable so conversion to a list
    stuff[0] = 0 # change value to 0 and index[0]
    for i in stuff:
        sum = sum + i
    return sum
print(add(1, 2, 3, 4, 5, 6))
```

# 36 . Game: Rock, Paper, Scissors


```python
import random

# variables for points
p_points = 0
c_points = 0

# whole game loop
while True:
    choices = ["rock", "paper", "scissors"]
    
    # variables for choices
    computer = random.choice(choices)
    player = None
    
    # while player input not in choices[] comes back to player input
    while player not in choices:
        player = input("rock, paper or scissors?: ").lower()

    # game rules
    if player == computer:
        p_points += 1
        c_points += 1
        print("computer: ", computer)
        print("player: ", player)
        print("Tie!")
    
    elif player == "rock":
        if computer == "paper":
            c_points += 1
            print("computer: ", computer)
            print("player", player)
            print("You loose!")
        if computer == "scissors":
            p_points += 1
            print("computer: ", computer)
            print("player", player)
            print("You win!")
    
    elif player == "scissors":
        if computer == "rock":
            c_points += 1
            print("computer: ", computer)
            print("player", player)
            print("You loose!")
        if computer == "paper":
            p_points += 1
            print("computer: ", computer)
            print("player", player)
            print("You win!")
    
    elif player == "paper":
        if computer == "scissors":
            c_points += 1
            print("computer: ", computer)
            print("player", player)
            print("You loose!")
        if computer == "rock":
            p_points += 1
            print("computer: ", computer)
            print("player", player)
            print("You win!")
    
    # play again ?
    play_again = input("Play again (y//n)?: ").lower()
    if play_again != "y":
        break

# end statement
print("Bye\nComputer points: {}\nPlayer points: {}".format(c_points, p_points))

if c_points > p_points:
    print("You loose!")
else:
    print("You win!")
```

# 37 . Game: Quiz

```python
# game rules function
def new_game():
    guesses = []
    correct_guesses = 0
    question_num = 1

    for key in questions:
        print("----------------------------------------------")
        print(key)
        for i in options[question_num - 1]:
            print(i)

        guess = input("Enter (A, B, C, D): ").upper()
        guesses.append(guess)
        
        correct_guesses += check_answer(questions.get(key), guess)
        question_num += 1

    display_score(correct_guesses, guesses)

# check for answer function
def check_answer(answer, guess):
    if answer == guess:
        print("correct!")
        return 1
    else:
        print("wrong!")
        return 0

# display score function
def display_score(correct_guesses, guesses):
    print("----------------------------------------------")
    print("RESULTS")
    print("----------------------------------------------")
    
    print("Answers: ", end="")
    for i in questions:
        print(questions.get(i), end=" ")
    print()

    print("Guesses: ", end="")
    for i in guesses:
        print(i, end=" ")
    print()

    score = int((correct_guesses / len(questions))*100)
    print("Your score is: " + str(score) + "%") 

# play again function
def play_again(): 
    response = input("Do you want to play again (y/n) ?: ").upper()
    if response == "Y":
        return True
    else:
        return False

# dictionary of questions and answers
questions = {"Who created Python?: ": "A",
             "What year Python was created?: ": "B",
             "Python is tributed to which comedy group?: ": "C",
             "Is the Earth round?: ": "A"}
# list of options in lists
options = [["A. Guido von Rossum", "B. Elon Musk", "C. Steve Jobs", "D. Mark Zuckerburg" ],
           ["A. 1989", "B. 1991", "C. 2000", "D. 2016"],
           ["A. Lonely Island", "B. Smosh", "C. Monty Python", "D. SNL"],
           ["A. True", "B. False", "C. Sometimes", "D. What's Earth?"]]

# game loop
new_game()
while play_again():
    new_game()
print("Bye!")
```

# 38 . Object oriented programming

```python
# Class can function as a blueprint to create objects
# We can assign attributes - what the object is or has
# and methods what each object can do.

# Within our class we have a special method called the
# __init__ method, we can pass in some arguments and
# assign these arguments to each object attributes.

# We can reuse this class as if it was a blueprint
# so we can create more objects out of it.

class Car:

    def __init__(self, make, model, year, color):
        self.make = make
        self.model = model
        self.year = year
        self.color = color

    def drive(self):
        print("This " + self.model + " is driving")

    def stop(self):
        print("This " + self.model + " is stopped")
```

# 39 . Class variables

```python
class Car:
    # you can set some default values outside of the method
    wheels = 4 # class variable
    
    # by default all cars will have 4 wheels
    def __init__(self, make, model, year, color):
        self.make = make    # instance variable
        self.model = model  # instance variable
        self.year = year    # instance variable
        self.color = color  # instance variable

# instance variable = declared inside of a constructor 
#                     __init__ they can be given unique values

# class variable = declared within the class but outside
#                  of a constructor __init__, you can set
#                  a default value for all instances of this class
#                  for all unique objects that are created and
#                  you can change its values too 
```

# 40 . Inheritance

```python
from re import A
# create class (classes can have children)
class Animal:

    alive = True

    def eat(self):
        print("This animal is eating")

    def slumber(self):
        print("This animal slumbers")

# create classes that inherits from previous class
class Rabbit(Animal):

    def run(self):
        print("This rabbit is running")


class Fish(Animal):
    def swim(self):
        print("This fish is swimming")


class Hawk(Animal):
    
    def fly(self):
        print("This hawk is flying")

rabbit = Rabbit()
fish = Fish()
hawk = Hawk()

print(rabbit.alive)
fish.eat()
hawk.slumber()

rabbit.run()
fish.swim()
hawk.fly()
```

# 41 . Multilevel inheritance:

```python
# multi-level inheritance = when a derived (child) class inherits 
#                           another derived (child) class

class Organism:

    alive = True

class Animal(Organism):
    def eat(self):
        print("This animal is eating")

class Dog(Animal):
    def bark(self):
        print("This dog is barking")

dog = Dog()

print(dog.alive)
dog.bark()
dog.eat()
```

# 42 . Multiple inheritance

```python
# multiple inheritance = when a child class is derived 
#                        from more than one parent class

class Prey:
    def flee(self):
        print("This animal flees")

class Predator:
    def hunt(self):
        print("This animal is hunting")


class Rabbit(Prey):
    pass

class Hawk(Predator):
    pass

class Fish(Prey, Predator):
    pass

rabbit = Rabbit()
hawk = Hawk()
fish = Fish()


rabbit.flee()

hawk.hunt()

print()

fish.flee()
fish.hunt()
```

# 43 . Method overriding



```python
# method overriding = ability of an object oriented programming language 
#                     to allow subclass also known as a child class to provide
#                     specific implementation of a method that is already
#                     provided by one of its parents.

# In this case we're going to override the eat method
class Animal:

    def eat(self):
        print("This animal eats")

class Rabbit(Animal):
    def eat(self):
        print("This rabbit is eating.")

rabbit = Rabbit()
rabbit.eat()
```
