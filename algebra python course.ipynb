# 01.Set up a proportion:

*Mnożenie ukośne:*  
$\frac{n1}{d1}=\frac{n2}{d2} $ to samo co:
$ n1*d2 = d1*n2 $

````python
# Put a zero in for the unknown value (x)
n1 = 1
d1 = 2
n2 = 4
d2 = 0

if n2 == 0:
    answer = d2 * n1 / d1
    print('n2 = ', answer)

if d2 == 0:
    answer = d1 * n2 / n1
    print('d2 = ', answer)
````

````python
import sympy
from sympy import symbols
from sympy.solvers import solve

x = symbols('x')

# Put the equation here
eq = 2*x - 4

print('x = ', solve(eq,x))

````

Doing more with the solution

````python
import sympy
from sympy import symbols
from sympy.solvers import solve

x = symbols('x')

# Put the equation here
eq = 2*x - 4

solution = solve(eq,x)
print('x = ', solution[0]) # [0] oznacza pierwsze rozwiązanie,2x-4 ma jedno
````

Multiple answers

````python
import sympy
from sympy import symbols
from sympy.solvers import solve

x = symbols('x')

# Two different ways to get the equation

eq = input('Enter equation: 0 = ')

solution = solve(eq,x)
for s in solution:  # Zmienna s dla wszystkich rozwiązań
    print('x = ', s)
````

Explaining how each function works


````python
# converts string input (including fractions) to float
def string_frac(in_string):
    if '/' in in_string:
        nd = in_string.split('/')
        n = float(nd[0])
        d = float(nd[1])
        answer = n / d
        return answer
    else:
        answer = float(in_string)
        return answer

# Simplest one-step addition
def one_step_add():
    import random
    a = random.randint(-4, 10)
    b = random.randint(2, 24)
    print('x + ', a, ' = ', b)
    ans = float(input('x = '))
    answer = b - a
    # Test input
    if ans==answer:
        print('Correct! \n')
    else:
        print('Try again')
        print('The correct answer is ', answer, '\n')

# One-step addition with negative numbers
def one_step_subtract():
    import random
    a = random.randint(-19, -1)
    b = random.randint(2, 24)
    print(a, '+ x = ', b)
    ans = float(input('x = '))
    answer = b - a
    # Test input
    if ans==answer:
        print('Correct! \n')
    else:
        print('Try again')
        print('The correct answer is ', answer, '\n')

# One-step multiply
def one_step_mult():
    # Uses string_frac(<input string>)
    import random
    a = random.randint(1, 11)
    b = random.randint(2, 24)
    print(a, 'x = ', b)
    ans_in = float(input('x = '))
    answer = b / a
    # Test input
    if string_frac(ans_in)==answer:
        print('Correct! \n')
    else:
        print('Try again')
        print('The correct answer is ', answer, '\n')
        
````
